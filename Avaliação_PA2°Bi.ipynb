#QUESTÃO 02 – Soma dos Números (2 pontos)
* Solicite ao usuário um número inteiro positivo.

Utilizando um laço de repetição, calcule e exiba a soma de
todos os números de 1 até o número informado.

Exemplo:

Digite um número: 5

Resultado: 15

Pois:

1 + 2 + 3 + 4 + 5 = 15

```python
#Q2
cnt = (int(input("Digite o número a ser contado: ")))
soma = 0
for i in range(cnt+1):
  soma = soma + i
  print(soma)
```

#QUESTÃO 04 – Fatorial (4 pontos)
* Solicite ao usuário um número inteiro positivo e calcule
seu fatorial utilizando um laço de repetição.

Exemplo:

Digite um número: 5

Fatorial = 120

Pois:

5! = 5 × 4 × 3 × 2 × 1

Observação:
O programa deve validar se o número informado é positivo.

```python
#Q4
a = (int(input("Digite um número a ser fatorado: ")))
fatorial = 1
for i in range(1,a+1):
  fatorial = fatorial * i
if a < 0:
  print("Número inválido")
if a > 0:
  print(f"O fatorial de {a} é {fatorial}")
```

#QUESTÃO 05 – Caixa Eletrônico (5 pontos)
* Um caixa eletrônico possui notas de:

* R$ 100

* R$ 50

* R$ 20

* R$ 10

* R$ 5

* R$ 2

* R$ 1

Solicite ao usuário o valor do saque e informe a menor
quantidade possível de notas para realizar o pagamento.

Exemplo:

Digite o valor do saque: 287

Saída esperada:

* 2 nota(s) de R$100

* 1 nota(s) de R$50

* 1 nota(s) de R$20

* 1 nota(s) de R$10

* 1 nota(s) de R$5

* 1 nota(s) de R$2

Requisitos:
- Utilizar laços de repetição.
- Não utilizar listas.
- Exibir apenas as notas utilizadas.

```python
#Q5
saque = int(input("Digite o valor do saque: "))
notas = [100, 50, 20, 10, 5, 2, 1]
for nota in notas:
  quantidade = saque // nota
  saque = saque % nota
  if quantidade > 0:
    print(f"{quantidade} nota(s) de R$ {nota}")
```
